```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='PREM',org=(50,200), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(8,9))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((3, 3), dtype=np.uint8)
```

```python
## PREM R
## 212223240124
```

```python
kernel
```

```python
erosion = cv2.erode(img, kernel, iterations=1)
```

```python
display_img(erosion)
```
